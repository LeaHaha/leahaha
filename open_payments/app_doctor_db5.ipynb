### Deleting Duplicates w.r.t State/City

```python
states = []
city_same = []
city_diff = []
with open('names_db5.csv','r+') as namelist:
    r = csv.reader(namelist)
    for i in r:
        if i[12] == 'mailing_state':
            states.append(i)
            city_same.append(i)
            city_diff.append(i)
        elif i[12] == '' or i[13] == '':
            states.append(i)
        elif i[12] == i[13]:
            if i[10].upper() == i[11]:
                city_same.append(i)
            else:
                city_diff.append(i)
                
newstate = pd.DataFrame(states) #At least one state field is null
newcitys = pd.DataFrame(city_same)#State field non-null, city matches
newcityd = pd.DataFrame(city_diff)#State field null, city doesn't match
```

```python
newstates = newstate.drop([0,1],1)
newstates.columns = ['first_last', 'middle_name', 'Physician_Profile_Middle_Name', 'suffix', 'Physician_Profile_Suffix_Name', 'full_name', 'npi', 'Physician_Profile_ID', 'mailing_city', 'Physician_Profile_City', 'mailing_state', 'Physician_Profile_State', 'mailing_zip', 'Physician_Profile_Zip_Code','Physician_ID']
newstates = newstates.drop(newstates.index[0])
empty_state = newstates.groupby('npi').filter(lambda x: x.count() == 1)
```

```python
newcityss = newcitys.drop([0,1],1)
newcityss.columns = ['first_last', 'middle_name', 'Physician_Profile_Middle_Name', 'suffix', 'Physician_Profile_Suffix_Name', 'full_name', 'npi', 'Physician_Profile_ID', 'mailing_city', 'Physician_Profile_City', 'mailing_state', 'Physician_Profile_State', 'mailing_zip', 'Physician_Profile_Zip_Code','Physician_ID']
newcityss = newcityss.drop(newcityss.index[0])
city_yes = newcityss.groupby('npi').filter(lambda x: x.count() == 1)

newcityds = newcityd.drop([0,1],1)
newcityds.columns = ['first_last', 'middle_name', 'Physician_Profile_Middle_Name', 'suffix', 'Physician_Profile_Suffix_Name', 'full_name', 'npi', 'Physician_Profile_ID', 'mailing_city', 'Physician_Profile_City', 'mailing_state', 'Physician_Profile_State', 'mailing_zip', 'Physician_Profile_Zip_Code','Physician_ID']
newcityds = newcityds.drop(newcityds.index[0])
city_no = newcityds.groupby('npi').filter(lambda x: x.count() == 1)
```

```python
city_match = city_yes.append(city_no)
citymatch = city_match.sort(['npi'])
```

```python
unicity = citymatch.groupby('npi').filter(lambda x: x.count() == 1)
twicity = citymatch.groupby('npi').filter(lambda x: x.count() > 1)
#twicity.to_csv('twincity.csv')
```

```python
sincity = twicity.loc[twicity['mailing_city'].map(lambda x: x.upper()) == twicity['Physician_Profile_City']]
fullcity = unicity.append(sincity)
fullstate = empty_state.append(fullcity)
```

```python
docmap = fullstate.groupby('npi').filter(lambda x: x.count() == 1)
cc = fullstate.groupby('npi').filter(lambda x: x.count()>1)
dupdoc = cc.loc[cc['mailing_state'].map(lambda x: x.upper()) == cc['Physician_Profile_State']]
```
